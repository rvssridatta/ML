# 2)plot for multiple lines

```python
years = range(2000,2005)
apples = [0.895, 0.91, 0.919, 0.926, 0.931]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('crops yields')
plt.legend(['apples', 'oranges']);
```

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('crops yields')
plt.legend(['apples', 'oranges']);
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('crops yields')
plt.legend(['apples', 'oranges']);
```

```python
plt.plot(years, apples, marker='o', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')  # one more short hand fmt = 's-b'
plt.plot(years, oranges, marker='x', c='r', ls='--', lw=3, ms=10, mew=10, alpha = 0.5) # fmt = 'x--r'
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('crops yields')
plt.legend(['apples', 'oranges']);
```

```python
plt.figure(figsize=(8,4))
plt.plot(years, oranges, 'or')
plt.title('yield of oranges');
```

# 6) multiple histograms

```python
flowers_df[flowers_df.sepal_width==3.3]
```

```python
setosa_df = flowers_df[flowers_df.species=='setosa']
versicolor_df = flowers_df[flowers_df.species=='versicolor']
virginica_df = flowers_df[flowers_df.species=='virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
```

```python
plt.title("Distribution of sepal width")
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width]
         ,bins = np.arange(2,5,0.25),
         stacked=True);

plt.legend(['setosa','versicolor','virginica'])
```

# 8) bar plots with averages

```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df
```

```python
bill_avg_df = tips_df.groupby('day')[['total_bill']].mean()
bill_avg_df
```

```python
plt.bar(bill_avg_df.index, bill_avg_df.total_bill)
```

```python
sns.barplot(data = tips_df, x ='day', y ='total_bill');
```

```python
sns.barplot(data = tips_df, x ='day', y ='total_bill', hue='sex')
```
