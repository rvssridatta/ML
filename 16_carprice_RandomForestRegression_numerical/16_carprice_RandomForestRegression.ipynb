```python
dataset = pd.read_csv('dataset.csv')
```

```python
dataset = dataset.drop(['car_ID'],axis = 1)
```

```python
dataset.head(5)
```

```python
dataset.shape
```

```python
#condsider only numerical cols
```

```python
Xdata = dataset.drop('price',axis = 'columns')
```

```python
numericalCols = Xdata.select_dtypes(exclude = ['object']).columns
X = Xdata[numericalCols]
X #includes only numercial cols
```

```python
Y = dataset['price']
Y
```

```python
from sklearn.preprocessing import scale
cols = X.columns
X = pd.DataFrame(scale(X))
X.columns = cols
X
```

```python
#splitting the data

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.20,random_state = 0)
```

```python
#training usng random forest

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
#evaluating model

ypred = model.predict(X_test)

from sklearn.metrics import r2_score
r2score = r2_score(y_test,ypred)
print('R2Score',r2score*100)
```
